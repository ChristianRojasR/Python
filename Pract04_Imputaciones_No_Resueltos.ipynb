Los datos faltantes pueden ocurrir cuando no se proporciona información para uno o más elementos o para una unidad completa. La falta de datos es un problema muy grande en el escenario de la vida real. Los datos faltantes también pueden denominarse NAvalores(no disponibles) en pandas. En DataFrame, a veces, muchos conjuntos de datos simplemente llegan con datos faltantes, ya sea porque existen y no se recopilaron o nunca existieron. Por ejemplo, supongamos que diferentes usuarios encuestados pueden optar por no compartir sus ingresos, algunos usuarios pueden optar por no compartir la dirección de esta manera, muchos conjuntos de datos desaparecieron.

En Pandas, los datos faltantes están representados por dos valores:

None: None es un objeto singleton de Python que se usa a menudo para los datos faltantes en el código de Python.

NaN: NaN(un acrónimo de Not a Number), es un valor de punto flotante especial reconocido por todos los sistemas que utilizan la representación de punto flotante estándar IEEE
Los pandas tratan Noney NaNcomo esencialmente intercambiables para indicar valores faltantes o nulos. 

Para facilitar esta convención, existen varias funciones útiles para detectar, eliminar y reemplazar valores nulos en Pandas DataFrame:

-isnull()

-notnull()

-dropna()

-fillna()

-reemplazar()

-interpolar()

Para verificar los valores faltantes en Pandas DataFrame, usamos una función isnull() y notnull(). Ambas funciones ayudan a comprobar si un valor lo es NaNo no. Esta función también se puede utilizar en Pandas Series para encontrar valores nulos en una serie.


Para verificar valores nulos en Pandas DataFrame, usamos la funcion isnull()

Esta función devuelve el marco de datos de valores booleanos que son verdaderos para los valores NaN.


````python
datos = {'First Score':[100, 90, np.nan, 95], 
        'Second Score': [30, 45, 56, np.nan], 
        'Third Score':[np.nan, 40, 80, 98]}

df = pd.DataFrame(datos)
df_bool = df.isnull()
df_bool
````
